```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
class graph:
    def __init__(self,xval,yval):
        self.xval=xval
        self.yval=yval
    def scatter(self,color,title):
        plt.plot(self.xval,self.yval,'o',color=color)
        plt.title(title)
        plt.xlabel("x value")
        plt.ylabel("y value")
    def line(self,color,title):
        plt.plot(self.xval,self.yval,color=color)
        plt.title(title)
        plt.xlabel("x value")
        plt.ylabel("y value")
    def dot(self,color,title):
        plt.plot(self.xval,self.yval,'.',color=color)
        plt.title(title)
        plt.xlabel("x value")
        plt.ylabel("y value")

FirstClu=graph(np.linspace(0,10,10),np.linspace(0,10,10))
        
```

```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
class graph3D:
    def __init__(self,xval,yval,zval):
        self.xval=xval
        self.yval=yval
        self.zval=zval
    def three_d(self,angle):
        fig = plt.figure(figsize=(8,5))
        ax = Axes3D(fig)
        ax.plot_surface(self.xval,self.yval, self.zval, rstride=1, cstride=1, cmap=mpl.cm.jet)
        ax.set_xlabel('\n\nX')
        ax.set_ylabel('    \n Y')
        ax.set_zlabel('Z')
        ax.view_init(30, angle[0])
        plt.show()
        
        fig = plt.figure(figsize=(8,5))
        ax = Axes3D(fig)
        ax.plot_surface(self.xval,self.yval, self.zval, rstride=1, cstride=1, cmap=mpl.cm.jet)
        ax.set_xlabel('\n\nX')
        ax.set_ylabel('    \n Y')
        ax.set_zlabel('Z')
        ax.view_init(30, angle[1])
        plt.show()
        
        fig = plt.figure(figsize=(8,5))
        ax = Axes3D(fig)
        ax.plot_surface(self.xval,self.yval, self.zval, rstride=1, cstride=1, cmap=mpl.cm.jet)
        ax.set_xlabel('\n\nX')
        ax.set_ylabel('    \n Y')
        ax.set_zlabel('Z')
        ax.view_init(30, angle[2])
        plt.show()
```
